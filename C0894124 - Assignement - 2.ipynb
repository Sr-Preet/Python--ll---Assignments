```python
# Q1
class Vehicle:
    def __init__(self, vehicle_name, max_speed, vehicle_average):
        self.vehicle_name = vehicle_name
        self.max_speed = max_speed
        self.vehicle_average = vehicle_average
        
    def display_vehicle_details(self):
        print(f'You have {self.vehicle_name}, which have max speed of {self.max_speed} and average of {self.vehicle_average}')
        
        
v1 = Vehicle('Toyota Corolla', 120, 25) 
v2 = Vehicle('Rubicon', 150, 20)

v1.display_vehicle_details()
v2.display_vehicle_details()

#Q2
class Car(Vehicle):
    def __init__(self,vehicle_name, max_speed, vehicle_average, vehicle_seating_capacity):
        super().__init__(vehicle_name, max_speed, vehicle_average)
        self.vehicle_seating_capacity = vehicle_seating_capacity
        
    def vehicle_seating_capacity_name(self, capacity):
        print(f"{self.vehicle_name} have seating capacity of {capacity}")
c1 = Car('Toyota Corolla', 120, 25, 4)

c1.vehicle_seating_capacity_name(4)


#Q3
# What is Inheritence?
'''
Inheritence is basically creating a child class that has its properties and methods but in 
addition to that it also inherits/derives properties and methods from it's parent class.

Multiple Inheritence is as the name says has not one class but two or more classes that a child class inherits from so
one can have properties/methods of multiple classes in child class.
'''

#Example of multiple Inheritence
class A:
    def A(self):
        print('This is class A.')        
class B:
    def B(self):
        print('This is class B.')        
class C(A,B):
    def C(self):
        print('This is class C which has multiple Inheritence and inherits features of both classes A and B.')
c1 = C()
c1.A()
c1.B()
c1.C()

#Q4
class Person:
    def __init__(self, name, age, gender):
        self.name = name
        self.__age = age
        self.gender = gender
        
    def say_hello(self):
        print(f'Person {self.name} says hello')

    def get_age(self):
        return self.__age
    
    def is_adult(self):
        if(self.get_age() >=18):
            return True
        else:
            return False
            
p1 = Person('John', 12, 'Male')
p2 = Person('Alice', 20, 'Female')
p1.say_hello()


if(p1.is_adult()):
    print(f'{p1.name} is adult')
else:
    print(f'{p1.name} is not adult')
    

class Student(Person):
    def __init__(self, student_id, course, name, age, gender):
        super().__init__(name, age, gender)
        self.student_id = student_id
        self.course = course
        
    def say_hello(self):
        print(f'Student {self.name} says hello')
        
    def student_course_details(self):
        print(f'{self.name} is enrolled in {self.course}')
        
s1 = Student(101, 'Full Stack Software Development', 'Dave', 18, 'Male')
s1.say_hello()
s1.student_course_details()

class Teacher(Person):
    def __init__(self, teacher_id, subject, name, age, gender):
        super().__init__(name, age, gender)
        self.teacher_id = teacher_id
        self.subject = subject
        
    def say_hello(self):
        print(f'Teacher {self.name} says Hello')
        
    def teacher_subject_details(self):
        print(f'{self.name} teaches {self.subject}')
        
        
t1 = Teacher(1001, 'Java', 'Alex', 35, 'Male')
t1.say_hello()
t1.teacher_subject_details()
```
